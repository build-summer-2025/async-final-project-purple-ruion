```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset. Replace 'titanic.csv' with your own dataset file for your project.
# Ensure the dataset file exists and is in the same directory as this jupyter notebook.
uber_df = pd.read_csv('ncr_ride_bookings.csv')

# Display the first few rows of the dataframe to ensure it is loaded correctly
display(uber_df)
```

```python
sns.countplot(x='Vehicle Type', data = uber_df)
plt.title('Count of Vehicle Types')
```

```python
plt.hist(x='Customer Rating', data=uber_df)
plt.title('Customer Ratings')
plt.ylabel('')
```

```python
num_no_driv = sum(uber_df['Booking Status'].str.contains('No Driver Found'))
num_incomplete = sum(uber_df['Booking Status'].str.contains('Incomplete'))
num_complete = sum(uber_df['Booking Status'].str.contains('Complete'))
num_can_cust = sum(uber_df['Booking Status'].str.contains('Cancelled by Customer'))
num_can_driv = sum(uber_df['Booking Status'].str.contains('Cancelled by Driver'))

num_of = [num_no_driv, num_incomplete, num_complete, num_can_cust, num_can_driv]
labels = ['No Driver', 'Incomplete', 'Complete', 'Cancelled by Customer', 'Cancelled by Driver']

plt.pie(num_of, labels=labels)
plt.axis('equal')
plt.title('Booking Status')

```

```python
num_cash = uber_df['Payment Method'].str.contains('Cash').sum()
num_uber_wallet = uber_df['Payment Method'].str.contains('Uber Wallet').sum()
num_upi = uber_df['Payment Method'].str.contains('UPI').sum()
num_cc = uber_df['Payment Method'].str.contains('Credit Card').sum()

payment_labels = ['Cash', 'Uber Wallet','UPI','Credit Card']
pay_num = [num_cash, num_uber_wallet, num_upi, num_cc]

plt.barh(payment_labels, pay_num)
plt.title('Payment Methods')
plt.xlabel('Count')
plt.ylabel('Payment Method')
```
