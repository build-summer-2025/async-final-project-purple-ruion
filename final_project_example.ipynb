```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset. Replace 'titanic.csv' with your own dataset file for your project.
# Ensure the dataset file exists and is in the same directory as this jupyter notebook.
uber_df = pd.read_csv('ncr_ride_bookings.csv')

# Display the first few rows of the dataframe to ensure it is loaded correctly
display(uber_df)
uber_df.info()
```

```python
#q1
sns.countplot(x='Vehicle Type', data = uber_df)
plt.title('Count of Vehicle Types')
```

```python
#q2
plt.hist(x='Customer Rating', data=uber_df)
plt.title('Customer Ratings')
plt.ylabel('Number of Ratings')
```

```python
#q3
num_no_driv = sum(uber_df['Booking Status'].str.contains('No Driver Found'))
num_incomplete = sum(uber_df['Booking Status'].str.contains('Incomplete'))
num_complete = sum(uber_df['Booking Status'].str.contains('Complete'))
num_can_cust = sum(uber_df['Booking Status'].str.contains('Cancelled by Customer'))
num_can_driv = sum(uber_df['Booking Status'].str.contains('Cancelled by Driver'))

num_of = [num_no_driv, num_incomplete, num_complete, num_can_cust, num_can_driv]
labels = ['No Driver', 'Incomplete', 'Complete', 'Cancelled by Customer', 'Cancelled by Driver']

plt.pie(num_of, labels=labels, autopct='%1.1f%%')
plt.axis('equal')
plt.title('Booking Status')

```

```python
#q4
num_cash = uber_df['Payment Method'].str.contains('Cash').sum()
num_uber_wallet = uber_df['Payment Method'].str.contains('Uber Wallet').sum()
num_upi = uber_df['Payment Method'].str.contains('UPI').sum()
num_cc = uber_df['Payment Method'].str.contains('Credit Card').sum()

payment_labels = ['Cash', 'Uber Wallet','UPI','Credit Card']
pay_num = [num_cash, num_uber_wallet, num_upi, num_cc]

plt.barh(payment_labels, pay_num)
plt.title('Payment Methods')
plt.xlabel('Count')
plt.ylabel('Payment Method')
```

```python
#q5
sample_df = uber_df.sample(100)

plt.scatter(sample_df['Driver Ratings'], sample_df['Customer Rating'])
plt.xlabel('Driver Ratings')
plt.ylabel('Customer Ratings')
```

```python
#q6 try line graph 
new_sample = uber_df.sample(100)

plt.scatter(new_sample['Ride Distance'], new_sample['Avg VTAT'])
plt.xlabel('Ride Distance')
plt.ylabel('VTAT')


```

```python
#what cities have the most rides across the dataset 
#try unique()
#print(uber_df['Driver Cancellation Reason'].unique())
#uber_df['Driver Cancellation Reason'].nunique()
#q7
r1 = uber_df['Driver Cancellation Reason'].str.contains('Personal & Car related issues').sum()
r2 = uber_df['Driver Cancellation Reason'].str.contains('Customer related issue').sum()
r3 = uber_df['Driver Cancellation Reason'].str.contains('More than permitted people in there').sum()
r4 = uber_df['Driver Cancellation Reason'].str.contains('The customer was coughing/sick').sum()

reasons = [r1,r2,r3,r4]
labels = ['Personal & Car related issues', 'Customer related issue', 'More than permitted people in there', 'The customer was coughing/sick']

plt.pie(reasons, labels=labels, autopct='%1.1f%%')
plt.axis('equal')
plt.title('Reasons why Driver Cancelled Ride')
```

```python
#q8
top5 = uber_df['Pickup Location'].value_counts().nlargest(5).index
top5_count = uber_df['Pickup Location'].value_counts().nlargest(5)
print(top5)

#sns.countplot(x='Pickup Location', data=uber_df, order=top5)
#plt.xticks(rotation=45, ha='right')
#plt.tight_layout()


plt.pie(top5_count, labels=top5, autopct='%.2f%%')
plt.title('Most Common Pickup Locations')
```

```python
#q9
khandsa_df = uber_df[uber_df['Pickup Location']=='Khandsa']

top5_drop = khandsa_df['Drop Location'].value_counts().nlargest(5).index
top5_c = khandsa_df['Drop Location'].value_counts().nlargest(5)

display(top5_c)
plt.pie(top5_c, labels=top5_drop, autopct='%.2f%%')
plt.title('Top Drop Locations in Khandsa')
```
